```python
from bs4 import BeautifulSoup
import pandas as pd
import xml.etree.ElementTree as et 
import os, glob
```

```python
def parse_XML(xml_file, df_cols): 
    
    xtree = et.parse(xml_file)
    xroot = xtree.getroot()
    rows = []
    
    for node in xroot: 
        res = []
        res.append(node.attrib.get(df_cols[0]))
        for el in df_cols[1:]: 
            if node is not None and node.find(el) is not None:
                res.append(node.find(el).text)
            else: 
                res.append(None)      
        rows.append({df_cols[i]: res[i] 
                     for i, _ in enumerate(df_cols)})
    
    out_df = pd.DataFrame(rows, columns=df_cols)
        
    return out_df
```

```python
# file to a variable under the name
# data
with open('q1eta1.wordlist.xml', 'r') as f:
    data = f.read()

# the beautifulsoup parser, storing
# the returned object
Bs_data = BeautifulSoup(data, "xml")
print (Bs_data)
```

```python
path = os.getcwd()
xml = glob.glob(os.path.join(path, "*.xml"))
trial = pd.DataFrame()
trial2 = pd.DataFrame()
trial3 = pd.DataFrame()
trial4 = pd.DataFrame()
t = pd.DataFrame()

for f in xml:
    file = parse_XML(f, ['id'])
    trial = trial.append(file)
    one = pd.concat([file, trial], axis='rows')
    
    file1 = parse_XML(f, ['start'])
    trial2 = trial2.append(file1)
    two = pd.concat([file1, trial2], axis='rows')
    
    file2 = parse_XML(f, ['end'])
    trial3 = trial3.append(file2)
    three = pd.concat([file2, trial3], axis='rows')
    
    file3 = parse_XML(f, ['type'])
    trial4 = trial4.append(file3)
    four = pd.concat([file3, trial4], axis='rows')

one.reset_index(drop=True, inplace=True)
two.reset_index(drop=True, inplace=True)
three.reset_index(drop=True, inplace=True)
four.reset_index(drop=True, inplace=True)

full = pd.concat([one, two, three, four], axis=1)
full = full.set_index('id')
full
```

```python
full.to_csv('wordlists.csv', sep=',', encoding='utf-8')
```
